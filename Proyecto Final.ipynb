```python
realestate.replace('nan', np.nan, inplace=True)
realestate['Residential Type'].unique()
```

```python
realestatetemp = realestate.isnull()
realestatetemp.head(10)
```

```python
for columns in realestatetemp.columns.values.tolist():
    print(realestatetemp[columns].value_counts())
    print('')
```

```python
realestatetemp = realestatetemp.dropna(subset=['Date Recorded'])
```

```python
for columns in realestatetemp.columns.values.tolist():
    print(realestatetemp[columns].value_counts())
    print('')
```

```python
realestatetemp['Date Recorded'].isna().sum()
```

```python
realestatetemp['Address'] = realestatetemp['Address'].fillna('ADDRESS NOT AVAILABLE')
```

```python
realestatetemp['Property Type'] = realestatetemp.groupby('Town')['Property Type'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
realestatetemp['Residential Type'] = realestatetemp.groupby('Town')['Property Type'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
mode_non_use_code = realestatetemp['Non Use Code'].mode()[0]
realestatetemp['Non Use Code'].fillna(mode_non_use_code, inplace=True)
```

```python
realestatetemp['Assessor Remarks'].fillna('NOT AVAILABLE', inplace=True)
```

```python
realestatetemp['OPM remarks'].fillna('NOT AVAILABLE', inplace=True)
```

```python
realestatetemp['Location'].fillna('NOT AVAILABLE', inplace=True)
```

```python
realestatetemp.update(realestate)
```

```python
realestatetemp = realestatetemp[~realestatetemp['Date Recorded'].isin([True, False])]
realestatetemp['Date Recorded'] = pd.to_datetime(realestatetemp['Date Recorded'], errors='coerce')
```

```python
realestatetemp['Date Recorded'] = pd.to_datetime(realestatetemp['Date Recorded'])
realestatetemp['List Year'] = realestatetemp['Date Recorded'].dt.year
annual_income = realestatetemp.groupby('List Year')['Sale Amount'].sum()
print(annual_income)
```

```python
realestatetemp['Sale Amount'] = pd.to_numeric(realestatetemp['Sale Amount'], errors='coerce')
```

```python
annual_income = realestatetemp.groupby('List Year')['Sale Amount'].sum()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
def dollar_formatter(x, pos):
    return f'${x:,.0f}'
plt.figure(figsize=(10,6))
annual_income.plot(kind='line', marker='o', linestyle='-', color='b')
plt.gca().yaxis.set_major_formatter(FuncFormatter(dollar_formatter))
plt.title('Ingreso Anual de Propiedades Vendidas por Año', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Ingreso Anual', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
residential_properties = realestatetemp[realestatetemp['Property Type'] == 'Residential']
commercial_properties = realestatetemp[realestatetemp['Property Type'] == 'Commercial']
residential_avg_price = residential_properties['Sale Amount'].mean()
commercial_avg_price = commercial_properties['Sale Amount'].mean()
print(f"Promedio Costo Propiedades Residenciales: ${residential_avg_price:,.2f}")
print(f"Promedio Costo Propiedades Comerciales: ${commercial_avg_price:,.2f}")
```

```python
import matplotlib.pyplot as plt
costs = [residential_avg_price, commercial_avg_price]
types = ['Residential', 'Commercial']
plt.figure(figsize=(8, 6))
bars = plt.barh(types, costs, color=['blue', 'green'])
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
             f'${bar.get_width():,.2f}', 
             va='center', ha='left', fontsize=12)
plt.title('Costo Promedio de Propiedades Residenciales y Comerciales', fontsize=14)
plt.xlabel('Costo Promedio ($)', fontsize=12)
plt.ylabel('Tipo de Propiedad', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
residential_properties = realestatetemp[realestatetemp['Property Type'] == 'Residential']
property_count = residential_properties['Residential Type'].value_counts()
city_sales = residential_properties.groupby(['Residential Type', 'Town'])['Sale Amount'].count().reset_index()
city_sales.columns = ['Residential Type', 'Town', 'Number of Sales']
avg_cost_per_type = residential_properties.groupby('Residential Type')['Sale Amount'].mean().reset_index()
avg_cost_per_type.columns = ['Residential Type', 'Average Sale Price']
merged_data = pd.merge(city_sales, avg_cost_per_type, on='Residential Type')
most_sold_property = property_count.idxmax()
most_sold_data = merged_data[merged_data['Residential Type'] == most_sold_property]
print(f"El tipo de propiedad residencial más vendido es: {most_sold_property}")
most_sold_city = most_sold_data.loc[most_sold_data['Number of Sales'].idxmax()]
print(f"La ciudad con más ventas de este tipo de propiedad es: {most_sold_city['Town']}")
print(f"El costo promedio de estas propiedades es: ${most_sold_city['Average Sale Price']:,.2f}")
plt.figure(figsize=(10, 6))
sns.barplot(x='Residential Type', y='Number of Sales', data=city_sales, palette='Blues_d')
plt.title('Ventas de Propiedades Residenciales por Tipo y Ciudad', fontsize=16)
plt.xlabel('Tipo de Propiedad Residencial', fontsize=12)
plt.ylabel('Número de Ventas', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
realestatetemp['Profit Margin'] = (realestatetemp['Sale Amount'] - realestatetemp['Assessed Value']) / realestatetemp['Assessed Value']
margin_by_property_type = realestatetemp.groupby('Property Type')['Profit Margin'].mean().reset_index()
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.barplot(x='Property Type', y='Profit Margin', data=margin_by_property_type, palette='viridis')
plt.title('Margen Promedio de Ganancia por Tipo de Propiedad', fontsize=16)
plt.xlabel('Tipo de Propiedad', fontsize=12)
plt.ylabel('Margen Promedio de Ganancia', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
